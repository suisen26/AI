#<font color="red">merge

```python
df1 = pd.DataFrame({'id': [1, 2, 3, 4, 5], 'item_id': ['A', 'B', 'C', None, 'G']})
df2 = pd.DataFrame({'id': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'], 'item': ['シャンプー', 'リンス', '石鹸', '洗顔', '歯磨き', '歯磨き粉', '洗剤', '柔軟剤']})
df1.merge(df2, left_on='item_id', right_on='id')
```

```python
df1.merge(df2, how='left', left_on='item_id', right_on='id')
```

```python
df1.merge(df2, how='right', left_on='item_id', right_on='id')
```
